```python
from PIL import Image
```

```python
import cv2
```

```python
image_to_ocr=cv2.imread('lawform.jpg')
```

```python
#step 1: Convert to gray scale
preprocessed_img=cv2.cvtColor(image_to_ocr,cv2.COLOR_RGB2GRAY)
```

```python
#STEP2: DO thresholding
preprocessed_img=cv2.threshold(preprocessed_img,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
```

```python
#STEP 3: Remove noise in image
preprocessed_img= cv2.medianBlur(preprocessed_img,3)
```

```python
#Save preprocessed image to convert to PIL image
cv2.imwrite("processed_regn.jpg",preprocessed_img)
```

```python
#load image as a PIL image
preprocessed_pil_img=Image.open('processed_regn.jpg')
```

```python
# Do OCR using tesseract
text_extracted = pytesseract.image_to_string(Image.open('processed_regn.jpg'))
```
